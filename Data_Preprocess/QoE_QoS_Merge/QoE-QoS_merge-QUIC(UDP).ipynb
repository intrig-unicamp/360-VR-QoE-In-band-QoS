### Normalize QoE values

```python
df1=pd.read_csv('/media/tariq/Data-House/vrexp/Datasets/new_quic-test.csv')
df1 = df1.drop(df1.index[df1['QoE'] < 0])

a, b = 0, 5
x, y = df1.QoE.min(), df1.QoE.max()
df1['nQoE'] = (df1.QoE - x) / (y - x) * (b - a) + a

x1, y1 = df1.QoE1.min(), df1.QoE1.max()
df1['nQoE1'] = (df1.QoE1 - x1) / (y1 - x1) * (b - a) + a


# x2, y2 = df1.TP_D_avg.min(), df1.TP_D_avg.max()
# df1['nTP_D_avg'] = (df1.TP_D_avg - x2) / (y2 - x2) * (b - a) + a

    
export_csv =df1.to_csv('/media/tariq/Data-House/vrexp/Datasets/new_quic-test1.csv', index=False)
```
